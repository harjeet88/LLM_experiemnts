```python
from google.colab import files
from gtts import gTTS
import IPython.display as ipd
```

```python
def text_to_speech(text, language='en', filename='output.mp3'):
    """
    Converts text to speech using Google Text-to-Speech (gTTS).

    Args:
        text (str): The text to convert to speech.
        language (str): The language code (e.g., 'en' for English, 'es' for Spanish).
        filename (str): The name of the output MP3 file.
    """
    try:
        tts = gTTS(text=text, lang=language)
        tts.save(filename)
        print(f"Audio saved as {filename}")
        ipd.display(ipd.Audio(filename)) #plays the audio in colab
        files.download(filename) #downloads the audio file.

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
text = """Say: 'It's a small issue.' Mean: 'The entire system is on fire, but I'm trying to stay calm.'

Just a quick fix.

Mean: 'I'll be here all night debugging this.'

The data is consistent.

Mean: 'The data is a chaotic mess, but I'm pretending it's fine.'

We'll have it deployed by EOD.

Mean: 'Maybe by the end of next week, if we're lucky.'

It's a legacy system.

Mean: 'It's held together by duct tape and prayers.'

Relatable? Subscribe to Data Savvy for more data engineering humor!
"""
```

```python
# Example usage:
text_to_speech(text, language='en', filename='test_audio.mp3')
```

```python
import os
```

```python
def generate_tts(text, filename, lang='en', slow=False):
    tts = gTTS(text=text, lang=lang, slow=slow)
    tts.save(filename)
    os.system(f"mpg123 {filename}") #Play the file. (Install mpg123 if needed)
```

```python
# Example with slight emphasis (using all caps for visual emphasis)
text1 = "This is a NORMAL voice."
text2 = "This is an EMPHASIZED voice!"
text3 = "This is a SLOW voice."
```

```python
generate_tts(text1, "normal.mp3")
generate_tts(text2, "emphasized.mp3")
generate_tts(text3, "slow.mp3", slow=True)
```
