```python
import torch
from transformers import BarkModel, AutoProcessor
```

```python
# Check for CUDA availability
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Load model and processor
processor = AutoProcessor.from_pretrained("suno/bark-small")
model = BarkModel.from_pretrained("suno/bark-small").to(device)
```

```python
# Text to generate speech from
text_prompt = "Hello, this is a test of Bark text-to-speech."
# Process the text
inputs = processor(text_prompt, return_tensors="pt").to(device)
```

```python
# Generate speech
with torch.no_grad():
    speech_output = model.generate(**inputs)
```

```python
# Convert to audio
sampling_rate = model.generation_config.sample_rate
audio_array = speech_output.cpu().numpy().squeeze()
```

```python
# Save as WAV file (requires soundfile library)
import soundfile as sf
sf.write("bark_output.wav", audio_array, sampling_rate)

print("Audio saved as bark_output.wav")
```
